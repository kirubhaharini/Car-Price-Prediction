```python
# Engine - convert to numeric

df['Engine']  = df['Engine'].str.replace(' cc', '')
df['Engine'] = pd.to_numeric(df['Engine'])
```

```python
print(df['Engine'].describe())
df['Engine'].hist()
#graph is skewed to the right - extreme outliers (max val) - so median filling used not mean
```

```python
# Filling missing Engine values with the median
df['Engine'] = df['Engine'].fillna(df['Engine'].median())
print(df['Engine'].isnull().sum())
df['Engine'].hist()
```

```python
# Repeat similar steps for other cols

# Clean 'Max Power' column (extract number)
df['Max Power (bhp)'] = df['Max Power'].str.extract(r'(\d*)').astype(float)

# Clean 'Max Torque' column (extract number)
df['Max Torque (Nm)'] = df['Max Torque'].str.extract(r'(\d*)').astype(float)
```

```python
print(df['Max Power (bhp)'].describe())
df['Max Power (bhp)'].hist()
```

```python
print(df['Max Torque (Nm)'].describe())
df['Max Torque (Nm)'].hist()
```

```python
df['Max Power (bhp)'] = df['Max Power (bhp)'].fillna(df['Max Power (bhp)'].median())
print(df['Max Power (bhp)'].isnull().sum())

df['Max Torque (Nm)'] = df['Max Torque (Nm)'].fillna(df['Max Torque (Nm)'].median())
print(df['Max Torque (Nm)'].isnull().sum())
```

```python
df.isnull().sum()
```

```python
print(df['Length'].describe())
df['Length'].hist()
```

```python
print(df['Width'].describe())
df['Width'].hist()
```

```python
print(df['Height'].describe())
df['Height'].hist()
```

```python
print(df['Fuel Tank Capacity'].describe())
df['Fuel Tank Capacity'].hist()
```

```python
df['Length'] = df['Length'].fillna(df['Length'].mean())
df['Width'] = df['Width'].fillna(df['Width'].mean())
df['Height'] = df['Height'].fillna(df['Height'].median()) #right skewed
df['Fuel Tank Capacity'] = df['Fuel Tank Capacity'].fillna(df['Fuel Tank Capacity'].median()) #right skewed
```

```python
#categorical values  

df['Seating Capacity'] = df['Seating Capacity'].fillna(df['Seating Capacity'].mode()[0])
df['Drivetrain'] = df['Drivetrain'].fillna(df['Drivetrain'].mode()[0])

#delete redundant cols
del df['Max Power']
del df['Max Torque']
```

### Encoding categorical values

```python
df

#encoding for categorical values

# Ordinal: Owner
# Nominal: Make (many), Model (many), Fuel Type, Transmission, Location (many), Color (many), Seller Type, Drivetrain

# -------------------- 1. Frequency Encoding for Model --------------------
df['Model'] = df['Model'].map(df['Model'].value_counts())

# -------------------- 2. Frequency Encoding for Location --------------------
df['Location'] = df['Location'].map(df['Location'].value_counts())

# -------------------- 3. Top-10 One-Hot Encoding for Make --------------------
top_makes = df['Make'].value_counts().nlargest(10).index
df['Make'] = df['Make'].apply(lambda x: x if x in top_makes else 'Other')

# -------------------- 4. Top-7 One-Hot Encoding for Color --------------------
top_colors = df['Color'].value_counts().nlargest(7).index
df['Color'] = df['Color'].apply(lambda x: x if x in top_colors else 'Other')

# -------------------- 5. Full One-Hot Encoding for Nominal Variables --------------------
df = pd.get_dummies(df, columns=[
    'Make',          # already grouped to top-N + Other
    'Color',         # already grouped to top-N + Other
    'Fuel Type',     # 9 unique, still manageable
    'Transmission',  # 2 unique
    'Seller Type',   # 3 unique
    'Drivetrain'     # 3 unique
], drop_first=True) # To remove redundancy - e.g. 0000 will be label A instead of creating a separate 0001 for it

# -------------------- 6. Label Encoding for Ordinal --------------------
le = LabelEncoder()
df['Owner'] = le.fit_transform(df['Owner']) 

```

### Preparation for ML

```python
print(df['Price'].describe())
df['Price'].hist()  #Right skewed - might need log transform
```

```python
y = np.log1p(df['Price']) #log transform Price (performs better) - since Price has very high values also
X = df.drop('Price', axis=1)
```

```python
X.shape, y.shape
```

```python
## Feature selection

selector = SelectKBest(score_func=f_regression,  k='all')
```

```python
# Try all features, plot scores

scores = selector.fit(X, y).scores_
plt.figure(figsize=(12,6))
plt.bar(range(len(scores)), scores)
plt.xticks(range(len(scores)), X.columns, rotation=90)
plt.title("Feature importance scores (f_regression)")
plt.tight_layout()
plt.show()
```

```python
selector = SelectKBest(score_func=f_regression,  k=9) # k = 9 based on plot
X_selected = selector.fit_transform(X, y)


selected_mask = selector.get_support()
selected_features = X.columns[selected_mask]
print(selected_features)
```

```python
X = df[selected_features]
```

```python
X.shape,y.shape
```
